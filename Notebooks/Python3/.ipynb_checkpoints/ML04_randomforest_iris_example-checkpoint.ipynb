<h3>Random Forest</h3>

<p>We can first try to see the difference when using this approach on the iris classification problem</p>

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data[:,0:3:2]
y = iris.target

from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
```

<p>Now training the classifier :</p>

```python
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X,y)
```

<p>Mapping the parameters space and visualizing the result... and check the difference</p>

```python
xx, yy = np.meshgrid(np.linspace(0.0,10.0,201), np.linspace(0.0,10.0,201))
pred_rfc = rfc.predict(np.c_[xx.ravel(),yy.ravel()])
Z_rfc = pred_rfc.reshape(np.shape(xx))
```

```python
plt.figure(figsize=(8,6))

plt.contourf(xx, yy, Z_rfc, alpha=0.7)
plt.scatter(X.T[0], X.T[1], c=y, edgecolors='k')
plt.axis('scaled')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Random Forest', fontsize=20)
plt.xlim(2.0,10.0)
plt.ylim(0.0,8.0)
```

<h4> NOTE : I have only call the algorithm with its default value, it is worth changing them (but think before doing any changes)</h4>

<h3>Extra (Randomized) Trees</h3>
<p>Just repeating hereafter the classification done with Random Forest but with the Extra Trees algorithm.</p>

```python
etr = ExtraTreesClassifier(n_estimators=100)
etr.fit(X,y)
```

```python
xx, yy = np.meshgrid(np.linspace(0.0,10.0,201), np.linspace(0.0,10.0,201))
pred_etr = etr.predict(np.c_[xx.ravel(),yy.ravel()])
Z_etr = pred_etr.reshape(np.shape(xx))
```

```python
plt.figure(figsize=(8,6))

plt.contourf(xx, yy, Z_etr, alpha=0.7)
plt.scatter(X.T[0], X.T[1], c=y, edgecolors='k')
plt.axis('scaled')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Extra Trees', fontsize=20)
plt.xlim(2.0,10.0)
plt.ylim(0.0,8.0)
```
